```python
def factorial(n):
    if n < 2:
        return 1
    else:
        return n * factorial(n-1)
```

```python
def funcion1(a,b):
    return a<b
def funcion2(a,b):
    return a>b
def funcion3(a,b):
    return a!=b

def buscar_valores(M,funcion1,threshold = 0):
    
    filas, columnas = M.shape
    
    for i in range(filas): # el contador i = 1, 2, 3, 4, 5
        for j in range(columnas): # el contador j= 1,2,...
            if funcion1(M[i,j],threshold):
                print("La posicion", (i+1,j+1))
```
